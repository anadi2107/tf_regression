```python
import pandas as pd
```

```python
data=pd.read_csv('cal_housing_clean.csv')
```

```python
data.head()
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(['medianHouseValue'],axis=1)
```

```python
y=data['medianHouseValue']
```

```python
X.head()
```

```python
y.head()
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=pd.DataFrame(data=scaler.transform(X_train),columns=X_train.columns,index=X_train.index)
```

```python
X_train.head()
```

```python
X_test=pd.DataFrame(data=scaler.transform(X_test),columns=X_test.columns,index=X_test.index)
```

```python
X_test
```

```python
data.columns
```

```python
import tensorflow.compat.v1 as tf
```

```python
age=tf.feature_column.numeric_column('housingMedianAge')
rooms=tf.feature_column.numeric_column('totalRooms')
bedrooms=tf.feature_column.numeric_column('totalBedrooms')
p=tf.feature_column.numeric_column('population')
h=tf.featrooms=tf.feature_column.numeric_column('households')
house=tf.feature_column.numeric_column('medianIncome')
```

```python
feat_ccols=[age,rooms,bedrooms,p,h,house]
```

```python
input_func=tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
model=tf.estimator.DNNRegressor(hidden_units=[6,6,6],feature_columns=feat_ccols)
```

```python
if tf.test.gpu_device_name(): 

    print('Default GPU Device:{}'.format(tf.test.gpu_device_name()))

else:

   print("Please install GPU version of TF")
```

```python
model.train(input_fn=input_func,steps=20000)
```

```python
predict_input_func=tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
pred_gen=model.predict(predict_input_func)
```

```python
predictions=list(pred_gen)
```

```python
predictions
```

```python
final_pred=[]
for pred in predictions:
    final_pred.append(pred['predictions'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,final_pred)**0.5
```
